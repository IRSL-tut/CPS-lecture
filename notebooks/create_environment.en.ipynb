# How to write environment file

https://github.com/IRSL-tut/irsl_sim_environments/blob/main/cnoid/world/depot_world.yaml
```yaml
world:
  World:
    name: DepotWorld
    draw_grid: False
  Camera:
    lookEye: [-5.5, -0.0, 2.0]
    lookUp: [0.0, 0.0, 1.0]
    lookAtCenter: [-2.5, -0.0, 0.7]
    fov: 0.75

objects:
  -
    model: 'package://irsl_sim_environments/cnoid/depot/depot.body'
    name: 'Depot'
    fix: True
  -
    model: 'package://irsl_sim_environments/cnoid/dining_table/dining_table.body'
    name: 'Table'
    initial_coords: { pos: [-2.5, 0.0, 0.0] }
    fix: True
```

## world (Top Item)
The following keywords can be set
- World
- Camera

### Configuration of World
- name: name of WorldItem
- draw_grid: drawing grid or not
- background: background color(not implemented yet)

### Configuration of Camera
- lookEye: Origin of camera
- lookUp: Up direction of image
- lookAtCener: Point through which the camera's optical axis passes

## objects (Top Item)
List of object

### Configuration of object
- model: file-name of body
- name: name of object
- initial_coords: initial coordinates (can read by irsl_choreonoid.robot_util.make_coordinates)
- fix: If True, this object is fixed to the environment
