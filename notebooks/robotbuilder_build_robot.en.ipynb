# Using RobotBuilder to make robot-model interactively

Import (IRSL standard)

```python
exec(open('/choreonoid_ws/install/share/irsl_choreonoid/sample/irsl_import.py').read())
```

## Make instance of RobotBuilder

```python
rb=RobotBuilder()
```

## Make root-link / Initial Link

making Box0

```python
b0=rb.makeBox(0.3, 0.5, 0.6, color=[0.2, 0.7, 0.2])
```

moving Box0

```python
b0.translate(fv(0,0,0.3))
```

making Box1

```python
b1=rb.makeBox(0.2, 0.2, 0.05, color=[0.2, 0.7, 0.2])
```

moving Box1

```python
b1.translate(fv(0,0.5/2 + 0.2/2,0.5))
```

Make a link from made geometries / createLinkFromShape

name: name of link

root: Set True, for making root-link

density: kg/m^3 Calculate weight/center of gravity/inertia with uniform density

```python
l_root=rb.createLinkFromShape(name='Root', root=True, density=400.0)
```

```python
%display
```

## Make joint (LINK0)

Making joint / createJointShape

jointType: Link.JointType.RevoluteJoint や Link.JointType.PrismaticJoint 

```python
j=rb.createJointShape(jointType=Link.JointType.RevoluteJoint)
```

Moving joint

```python
j.rotate(-PI/2, coordinates.Z)
j.locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5), coordinates.wrt.world)
```

Making geometries

```python
b2=rb.makeCylinder(0.06, 0.3, color=[0.7, 0.2, 0.2])
b2.rotate(-PI/2, coordinates.Z)
b2.locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5), coordinates.wrt.world)
b3=rb.makeBox(0.1, 0.1, 0.4, color=[0.2, 0.7, 0.2])
b3.locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4/2), coordinates.wrt.world)
```

Making a link from made geometries / createLinkFromShape 

Make joint where the geometry is displayed

parentLink: parent-link

JointId: Should be set for making it as controlled joint

JointName:

InitialJointAngle:

JointRange:

JointVelocityRange:

JointEffortRange:

EquivalentRotorInertia:


```python
l0=rb.createLinkFromShape(name='LINK0', parentLink=l_root, density=400.0, JointId=0, JointName='JOINT0', InitialJointAngle=0.6, JointRange=[0, PI], JointVelocityRange=[-PI*10, PI*10], JointEffortRange=[-100, 100], EquivalentRotorInertia=0.1)
```

```python
%display
```

## Make joint (LINK1)

Making joint and geometries

```python
rb.createJointShape(jointType=Link.JointType.RevoluteJoint).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4), coordinates.wrt.world)
rb.makeCylinder(0.06, 0.3, color=[0.7, 0.2, 0.2]).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4), coordinates.wrt.world)
rb.makeBox(0.1, 0.1, 0.4, color=[0.2, 0.7, 0.2]).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4-0.4/2), coordinates.wrt.world)
```

Note: Don't mistake parent link

Note: Set correct id to JointId

Since JointId is an index, it must be specified sequentially from 0 without skipping any numbers

```python
l1=rb.createLinkFromShape(name='LINK1', parentLink=l0, density=400.0, JointId=1, JointName='JOINT1', InitialJointAngle=0.6, JointRange=[-PI/2, PI/2], JointVelocityRange=[-PI*10, PI*10], JointEffortRange=[-100, 100], EquivalentRotorInertia=0.1)
```

```python
%display
```

## Make joint (LINK2)

Making prismatic-joint

```python
rb.createJointShape(jointType=Link.JointType.PrismaticJoint).rotate(-PI/2, coordinates.X).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4-0.4), coordinates.wrt.world)
rb.makeBox(0.08, 0.08, 0.3, color=[0.7, 0.2, 0.2]).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4-0.4+0.3/2), coordinates.wrt.world)
rb.makeBox(0.14, 0.14, 0.14, color=[0.2, 0.7, 0.2]).locate(fv(0, 0.5/2 + 0.2/2 + 0.2/2, 0.5-0.4-0.4-0.14/2), coordinates.wrt.world)
```

Making link

```python
l2=rb.createLinkFromShape(name='LINK2', parentLink=l1, density=400.0, JointId=2, JointName='JOINT2', InitialJointAngle=0.6, JointRange=[0, 0.3], JointVelocityRange=[10, 10], JointEffortRange=[-100, 100], EquivalentRotorInertia=0.1)
```
